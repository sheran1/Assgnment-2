```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Coffee_Shop': ['Brewed Awakening', 'Coffee Cloud', 'Bean Dream', 'Espresso Express', 'Latte Love',
    'Mocha Magic', 'Cafe Comfort', 'Seattle Sip', 'Drip Drop', 'Grind Ground'],
    'Location': ['Downtown', 'Capitol Hill', 'Green Lake', 'Ballard', 'West Seattle',
    'Fremont', 'Queen Anne', 'Belltown', 'University District', 'Magnolia'],
    'Avg_Rating': [4.5, 4.2, 5.0, 4.8, 4.6, 4.3, 4.9, 4.0, 4.7, 4.6],
    'Coffee_Type': ['Espresso', 'Latte', 'Cappuccino', 'Americano', 'Cold Brew',
    'Macchiato', 'Espresso', 'Latte', 'Drip Coffee', 'Americano'],
    'Tables_Available': [5, 8, 6, 7, 5, 9, 7, 8, 6, 5]
}
coffee_shops_df = pd.DataFrame(data)
coffee_shops_df
```

# How would you use the clip method to ensure that all Avg_Rating values in the coffee_shops_df are between 4.0 and 4.5

```python
temp = coffee_shops_df.copy()
temp.loc[:, 'Avg_Rating'] = coffee_shops_df.loc[:, 'Avg_Rating'].clip(4.0, 4.5)
temp
```

# If you wanted to clip only the lower bound of Tables_Available to 6 without changing the upper values, how would you do it?

```python
temp = coffee_shops_df.copy()
```

```python
temp.loc[:, 'Tables_Available'] = temp.loc[:, 'Tables_Available'].clip(lower=6)
```

```python
temp
```

# What happens to the original DataFrame after using the clip method? Is it modified in place or do you get a new DataFrame?

```python
#How would you clip all numerical columns in a DataFrame at once?
coffee_shops_df.select_dtypes(include=[np.number]).clip(4.4, 8)
```

# What's the difference between using the clip method and using conditional statements to limit data range?

```python
coffee_shops_df.loc[(coffee_shops_df['Avg_Rating'] < 4.0) | (coffee_shops_df['Avg_Rating'] > 4.4)]
```

```python
temp = coffee_shops_df.copy()
temp.loc[:, 'Avg_Rating'] = coffee_shops_df.loc[:, 'Avg_Rating'].clip(4.0, 4.4)
temp
```

# iterating_through_groups

# 1) How can you iterate through each group created by grouping the DataFrame by the 'Location' column, and print the name of each location along with the number of shops in that location?¶

```python
coffee_shops.groupby('Location').count()
```

```python
grouped = coffee_shops.groupby('Location')
```

```python
for name, group in grouped:
    print(f"Location '{name}': ", end='')
    print(f"{len(group)} shops")
    print()
```

```python
for name,  group in coffee_shops.groupby(['Type', 'Location']):
    print(f"Group - {name} : ", end="")
    print(group.loc[:,'Rating'].mean())
```

```python
type_grouped = coffee_shops.groupby('Type')
```

```python
for type_name, type_group in type_grouped:
    print(f"{type_name} : ", end="")
    cur_max_for_type = -math.inf
    for loc_name, loc_group in type_group.groupby('Location'):
        cur_loc_sum = loc_group.loc[:, 'Rating'].sum()
        if cur_loc_sum > cur_max_for_type:
            cur_max_for_type = cur_loc_sum
    print(cur_max_for_type)
```

# How many unique combinations of 'Product_ID' and 'Price' are there?

```python
prod_price_pair = df.loc[:, ['Product_ID', 'Price']]
```

```python
prod_price_pair
```

```python
def get_pair(row):
    return (row['Product_ID'], row['Price'])
```

```python
prod_price_pair_series = prod_price_pair.apply(get_pair, axis=1)
```

```python
prod_price_pair_series
```

```python
unique_pairs = pd.unique(prod_price_pair_series)
```

```python
unique_pairs
```

```python
pd.DataFrame(list(unique_pairs), columns=['Product_ID', 'Price'])
```

# Are the units in which products sold always unique, or do some products share the same number of 'Units_Sold'? Use nunique() to determine this

```python
df1 = df.copy()
```

```python
df1.loc[8, 'Units_Sold'] = 190 # to make one product duplicate
```

```python
df1
```

```python
df1.groupby('Product_Name')['Units_Sold'].sum()
```

```python
actual_items = len(df1.groupby('Product_Name')['Units_Sold'].sum())
```

```python
unique_items = df1.groupby('Product_Name')['Units_Sold'].sum().nunique()
```

```python
actual_items, unique_items
```
